Sony

```python
from bs4 import BeautifulSoup
import pandas as pd

# Cargar el archivo HTML (ajusta el nombre si es necesario)
with open("sony.html", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

# Marcas objetivo
marcas_objetivo = ["asus", "microsoft", "nintendo", "sony"]

# Lista de productos
productos = []

# Función para limpiar y convertir precios
def limpiar_precio(p):
    if p:
        return float(p.replace("S/", "").replace(",", "").strip())
    return None

# Scraping
for item in soup.select("li.product-item"):
    nombre_elem = item.select_one("a.product-item-link")
    precio_actual_elem = item.select_one("span.price")
    precio_anterior_elem = item.select_one("span.old-price .price")

    modelo = nombre_elem.get_text(strip=True) if nombre_elem else None
    precio_actual = limpiar_precio(precio_actual_elem.get_text()) if precio_actual_elem else None
    precio_anterior = limpiar_precio(precio_anterior_elem.get_text()) if precio_anterior_elem else None

    if modelo:
        marca = next((m.capitalize() for m in marcas_objetivo if m in modelo.lower()), None)
        if marca and precio_actual:
            descuento_soles = None
            descuento_pct = None

            if precio_anterior:
                descuento_soles = round(precio_anterior - precio_actual, 2)
                descuento_pct = round((descuento_soles / precio_anterior) * 100, 1)

            productos.append({
                "modelo": modelo,
                "marca": marca,
                "precio_actual": precio_actual,
                "precio_anterior": precio_anterior,
                "descuento_soles": descuento_soles,
                "descuento_porcentaje": descuento_pct
            })

# Crear DataFrame
df = pd.DataFrame(productos)

# Exportar CSV
df.to_csv("sony.csv", index=False, encoding="utf-8-sig")
print("✅ Archivo 'sony.csv' generado con descuentos.")

# Vista previa
df.head()
```
